```python
from math import sqrt, factorial
from time import time

t = time()
def prime_printer(n: int):
    """prints the primes till n"""

    lst = [True]*n
    p = []
    t = time()
    
    for i in range(2, int(sqrt(n))): 
        if lst[i]:
            for j in range(i**2, n, i):
                lst[j] = False

    for i in range(2, n): 
        if lst[i]: 
            p.append(i)

    return p

l = prime_printer(10000)
my_list = []

for i in l:
    if i % 8 == 5:
        my_list.append(i)

result = [[item_1,item_2] for item_1,
          item_2 in zip(my_list[::2], my_list[1::2])]

#print(my_list)
print(result)  


print(f'ended in {time()-t} s')
```
