```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return (x+1)**2

def L(x, l):
    return f(x) + l*(x+2)*(x-4)
    
def g(x):
    return (-9 * x**2 - 5*x) / (x+1)

def zero(x):
    return 0*x
```

```python
plt.figure(figsize=(12, 6))
plt.xticks([i for i in range(-2, 5)])
xs = np.linspace(-2, 4, 1000)
plt.plot(xs, f(xs), c="b", label="f(x)")
plt.plot(xs, L(xs, 1), c="g", label="L(x, 1)") 
plt.plot(xs, L(xs, 2), c="g", label="L(x, 2)") 
plt.plot(xs, L(xs, 3), c="g", label="L(x, 3)") 
plt.plot([-1], f(-1), 'bo', label="optimal value")
plt.plot(xs, zero(xs), c="k", linewidth=1.0)
plt.title("")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.xticks([i for i in range(0, 5)])
ls = np.linspace(0, 4, 1000)
plt.plot(ls, g(ls), c="r", label="g($\lambda$)")
plt.plot([0], g(0), 'ro', label="optimal value")
plt.plot(ls, zero(xs), c="k", linewidth=1.0)
plt.title("")
plt.legend()
plt.show()
```
